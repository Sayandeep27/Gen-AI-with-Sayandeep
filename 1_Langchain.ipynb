3️⃣ Free Hugging Face Models (via API or Locally)
Hugging Face provides two ways to use models:

Via API (needs free HF token but runs in the cloud).

Download & run locally (no internet needed, like method 2).

A) Using Hugging Face Models via API (Cloud-Based)
Uses models like mistralai/Mistral-7B-Instruct via API.

No GPU needed, runs on Hugging Face’s cloud servers.

```python
from langchain_huggingface import HuggingFaceEndpoint

llm = HuggingFaceEndpoint(
    repo_id="mistralai/Mistral-7B-Instruct",
    huggingfacehub_api_token="your-huggingface-token"
)

response = llm.invoke("Explain deep learning.")
print(response)
```

B) Running Hugging Face Models Locally
No API calls, runs directly on your machine.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_id = "facebook/bart-large-cnn"  # Change to any HF model

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(model_id)

# Create pipeline
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

response = pipe("What is AI?", max_length=100)
print(response)
```

Load a Free Model from Hugging Face but use Langchain as well

```python
from langchain_community.llms import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# ✅ Use a public model (NO token required)
model_name = "facebook/opt-1.3b"

# Load tokenizer & model locally
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto")

# Create a text-generation pipeline
text_pipeline = pipeline("text-generation", model=model, tokenizer=tokenizer, max_length=200)

# Integrate with LangChain
llm = HuggingFacePipeline(pipeline=text_pipeline)

# Test the model
response = llm.invoke("Explain machine learning in simple terms.")
print(response)
```
